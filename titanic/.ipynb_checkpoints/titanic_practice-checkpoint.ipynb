## Creating - catagorical variables from continuous variable

- We witnessed that age and fare might introduce orders of magnitude bias, so let's convert those continuous variables into categorical variables through the process of binning. 

1. For age, we are manually setting the ben limits, but what would be a good way to find the optimal set of bin limits.

    * At what age do we see people stop considering a person a child and thus their likelihood to survive is no longer higher than a teen or an adult. 
    * Are teens more likely to survive than young adults? how do you find the cut off between teens and young adults?
    * Are seniors more likely to survive than young adults?
    * Combine age with sex and do young women have better chances of surviving than old women?
    * Does a young adult woman have a similar chance of survival to a teen boy? 

```python
for dataset in datasets:
    dataset['GroupSize'] = dataset['SibSp'] + dataset['Parch'] + 1
```

```python
train_df['GroupSize'].min(),train_df['GroupSize'].max()
```

```python
bins = [0,1,3,5,8,12]
labels = ['Single','Couple','Small Group','Medium Group','Large Group']
for dataset in datasets:
    dataset['GroupType'] = pd.cut(dataset['GroupSize'].astype(int), bins, labels=labels)
```

```python
bins = [0,5,10,15,30,55,80]
labels = ['Baby','Child','Teen','Young Adult','Adult','Senior']
for dataset in datasets:
    dataset['AgeBin'] = pd.cut(dataset['Age'].astype(float), bins, labels=labels)
```

```python
for dataset in datasets:
    dataset['FareBin'] = pd.qcut(dataset['Fare'],5, labels=['Q1','Q2','Q3','Q4','Q5'])
```

```python
#correlation heatmap of dataset
def correlation_heatmap(df):
    _ , ax = plt.subplots(figsize =(14, 12))
    colormap = sns.diverging_palette(220, 10, as_cmap = True)
    
    _ = sns.heatmap(
        df.corr(), 
        cmap = colormap,
        square=True, 
        cbar_kws={'shrink':.9 }, 
        ax=ax,
        annot=True, 
        linewidths=0.1,vmax=1.0, linecolor='white',
        annot_kws={'fontsize':12 }
    )
    
    plt.title('Pearson Correlation of Features', y=1.05, size=15)

correlation_heatmap(train_df[target+numeric_columns+['AgeBin_Code','GroupType_Code','Sex_Code','FareBin_Code']])
```

```python
train_set = train_df[['Survived','Pclass','Sex_Code','Age','SibSp','Parch','Fare']]
test_set = test_df[['Pclass','Sex_Code','Age','SibSp','Parch','Fare']]

new_series = train_set["Survived"]
# Split and scale the data
X_train, X_test, y_train, y_test = cross_validation.train_test_split(train_set[train_set.columns.drop("Survived")], new_series, train_size=0.75)
scaler = preprocessing.StandardScaler()

X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test = scaler.transform(X_test)
# logistic regression
cls = linear_model.LogisticRegression()

cls.fit(X_train, y_train)
y_pred = cls.predict(X_test)
print accuracy_score(y_test, y_pred)
```

```python
coefs = pd.Series(cls.coef_[0], index=X_train.columns)
print X_train.columns
coefs = coefs.sort_values()
plt.subplot(1,1,1)
coefs.plot(kind="bar")
plt.show()
print coefs.sort_values(ascending = True)
```

## Observation

- Sex, Class, Age, and whether or not you were traveling with a spouse/sibling seem to be the best predictors of survival
